```python
import pandas as pd
import numpy as np
import sqlite3

from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score, classification_report
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect("inventory.db")
df = pd.read_sql("SELECT * FROM vendor_sales_summary", conn)

print(df.head())
```

```python
# Remove invalid values
df = df[df['TotalSalesDollars'] > 0]

# Remove extreme outliers
df = df[df['TotalSalesDollars'] < df['TotalSalesDollars'].quantile(0.99)]
```

```python
df['ProfitMargin'] = np.where(
    df['TotalSalesDollars'] != 0,
    df['GrossProfit'] / df['TotalSalesDollars'],
    0
)

df['ProfitMargin'] = df['ProfitMargin'].clip(-1, 1)
```

```python
def classify_vendor(margin):
    if margin > 0.3:
        return "High"
    elif margin > 0.1:
        return "Medium"
    else:
        return "Low"

df['VendorCategory'] = df['ProfitMargin'].apply(classify_vendor)
```

```python
reg_features = [
    'TotalPurchaseDollars',
    'StockTurnover',
    'SalesToPurchaseRatio'
]

X_reg = df[reg_features]
y_reg = df['TotalSalesDollars']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)
```

```python
reg_model = GradientBoostingRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=4,
    random_state=42
)

reg_model.fit(X_train, y_train)
```

```python
y_pred = reg_model.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse_percent = (rmse / y_test.mean()) * 100
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("RMSE %:", rmse_percent)
print("R2 Score:", r2)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted Sales")
plt.show()
```

```python
importance = pd.DataFrame({
    'Feature': reg_features,
    'Importance': reg_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

print(importance)
```

```python
clf_features = [
    'TotalSalesDollars',
    'TotalPurchaseDollars',
    'ProfitMargin',
    'StockTurnover',
    'SalesToPurchaseRatio'
]

X_clf = df[clf_features]
y_clf = df['VendorCategory']
```

```python
clf_model = RandomForestClassifier(n_estimators=100, random_state=42)
clf_model.fit(X_clf, y_clf)
```

```python
y_pred_clf = clf_model.predict(X_clf)

print(classification_report(y_clf, y_pred_clf))
```

```python
def predict_vendor(data):
    
    # Convert to DataFrame
    data = pd.DataFrame([data])
    
    # Predict Sales
    sales_pred = reg_model.predict(data[reg_features])[0]
    
    # Calculate Profit Margin
    purchase = data['TotalPurchaseDollars'].values[0]
    
    if sales_pred != 0:
        profit_margin = (sales_pred - purchase) / sales_pred
    else:
        profit_margin = 0
    
    # Prepare classification input
    clf_input = pd.DataFrame([{
        'TotalSalesDollars': sales_pred,
        'TotalPurchaseDollars': purchase,
        'ProfitMargin': profit_margin,
        'StockTurnover': data['StockTurnover'].values[0],
        'SalesToPurchaseRatio': data['SalesToPurchaseRatio'].values[0]
    }])
    
    # Predict Category
    category = clf_model.predict(clf_input)[0]
    
    return sales_pred, category
```

```python
new_vendor = {
    'TotalPurchaseDollars': 300000,
    'StockTurnover': 6,
    'SalesToPurchaseRatio': 1.5
}

sales, category = predict_vendor(new_vendor)

print("Predicted Sales:", sales)
print("Vendor Category:", category)
```
